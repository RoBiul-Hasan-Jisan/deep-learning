```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# === Function & Gradient ===
def f(x, y):
    return x**4 - x**2 + y**2  # Saddle surface

def grad_f(x, y):
    return 4 * x**3 - 2 * x, 2 * y

# === Gradient Descent ===
def gradient_descent(x0, y0, lr, epochs):
    x, y = x0, y0
    path = [(x, y, f(x, y))]
    losses = [f(x, y)]

    for _ in range(epochs):
        dx, dy = grad_f(x, y)
        x -= lr * dx
        y -= lr * dy
        path.append((x, y, f(x, y)))
        losses.append(f(x, y))
    
    return path, losses

# === Visualization Helpers ===
def plot_contour(path, title):
    x_vals, y_vals, _ = zip(*path)
    X = np.linspace(-2, 2, 400)
    Y = np.linspace(-2, 2, 400)
    X, Y = np.meshgrid(X, Y)
    Z = f(X, Y)

    plt.figure(figsize=(6,6))
    plt.contour(X, Y, Z, levels=50, cmap='viridis')
    plt.plot(x_vals, y_vals, marker='o', color='red', label='Path')
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.legend()
    plt.axis("equal")
    plt.show()

def plot_3d(path, title):
    x_vals, y_vals, z_vals = zip(*path)
    X = np.linspace(-2, 2, 100)
    Y = np.linspace(-2, 2, 100)
    X, Y = np.meshgrid(X, Y)
    Z = f(X, Y)

    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)
    ax.plot(x_vals, y_vals, z_vals, color='red', marker='o', label='Path')
    ax.set_title(title)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("f(x, y)")
    ax.legend()
    plt.show()

def plot_loss_curve(losses, title):
    plt.figure(figsize=(6, 4))
    plt.plot(losses, color='blue')
    plt.title(f"Loss Curve: {title}")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.grid(True)
    plt.show()
```

```python
#  Damped Sine Function and its Derivative

#This notebook explores the function:

''' [
f(x) = \sin(x) \cdot e^{-0.05x^2}
\]

and its analytical derivative:

\[
f'(x) = \cos(x) \cdot e^{-0.05x^2} - 0.1x \cdot \sin(x) \cdot e^{-0.05x^2}
\]

We will plot both the function and its derivative.'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return np.sin(x) * np.exp(-0.05 * x**2)

def df(x):
    return np.cos(x) * np.exp(-0.05 * x**2) - 0.1 * x * np.sin(x) * np.exp(-0.05 * x**2)

# Gradient Descent Function
def gradient_descent(f, df, x0, lr=0.05, epochs=100, track=True):
    x = x0
    history = [x]
    for _ in range(epochs):
        grad = df(x)
        x = x - lr * grad
        if track:
            history.append(x)
    return x, history

# Set up x range for plotting
X = np.linspace(-10, 10, 1000)
Y = f(X)
dY = df(X)

# Run Gradient Descent
x_init = 1.4
lr = 0.05
epochs = 100
final_x, traj = gradient_descent(f, df, x_init, lr=lr, epochs=epochs)

# Plot
plt.figure(figsize=(12, 6))

# Function and Derivative
plt.plot(X, Y, label="f(x) = sin(x) * exp(-0.05 x)")
plt.plot(X, dY, label="f'(x)", linestyle='--')

# GD Trajectory
traj_y = f(np.array(traj))
plt.plot(traj, traj_y, 'ro-', markersize=4, label='GD Path')
plt.scatter(traj[0], traj_y[0], color='green', label='Start', zorder=5)
plt.scatter(traj[-1], traj_y[-1], color='black', label='End', zorder=5)

plt.title(f"Gradient Descent on f(x), start={x_init}, lr={lr}, epochs={epochs}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
