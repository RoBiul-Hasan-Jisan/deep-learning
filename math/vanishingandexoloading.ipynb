```python
# Vanishing and Exploding Gradient Visualization 

import numpy as np
import matplotlib.pyplot as plt

# Function to simulate gradient flow over layers
def simulate_gradient_flow(multiplier, steps=50):
    grad = 1.0  # initial gradient magnitude (scalar)
    grads = []
    for _ in range(steps):
        grad *= multiplier  # simulate gradient scaling through layers
        grads.append(abs(grad))
    return grads

# Parameters
steps = 50
vanishing_factor = 0.5    # gradient shrinks each layer → vanishing gradient
exploding_factor = 1.5    # gradient grows each layer → exploding gradient

# Simulate
vanishing_grads = simulate_gradient_flow(vanishing_factor, steps)
exploding_grads = simulate_gradient_flow(exploding_factor, steps)

# Plot results
plt.figure(figsize=(10,6))
plt.plot(vanishing_grads, label="Vanishing Gradient (multiplier=0.5)", color='blue')
plt.plot(exploding_grads, label="Exploding Gradient (multiplier=1.5)", color='red')
plt.title("Vanishing vs Exploding Gradient Illustration")
plt.xlabel("Layer (Step)")
plt.ylabel("Gradient Magnitude")
plt.grid(True)
plt.legend()
plt.show()

# Explanation
print("""
Vanishing Gradient:
- Gradient magnitude shrinks exponentially as it flows backward through layers.
- Makes early layers learn very slowly or not at all.

Exploding Gradient:
- Gradient magnitude grows exponentially, causing unstable updates.
- Can make training diverge or produce NaN values.

Common fixes include using ReLU activations, normalization techniques, 
skip connections (ResNets), gradient clipping, and careful weight initialization.
""")
```
