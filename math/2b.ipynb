```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def f(x, y):
    return x**2 + y**2

# Gradient of the function
def grad_f(x, y):
    return 2*x, 2*y

# Initialize
x, y = 4.0, 3.0  # Starting point
lr = 0.1         # Learning rate
steps = 30

# Store the path
path_x, path_y, path_z = [x], [y], [f(x, y)]

# Gradient descent loop
for _ in range(steps):
    dx, dy = grad_f(x, y)
    x -= lr * dx
    y -= lr * dy
    path_x.append(x)
    path_y.append(y)
    path_z.append(f(x, y))

# Plot the surface and path
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)
ax.plot(path_x, path_y, path_z, color='red', marker='o')
ax.set_title("2D Gradient Descent Path")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Loss")
plt.show()
```
