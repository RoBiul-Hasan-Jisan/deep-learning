```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def f(x, y):
    return x**2 + y**2

# Gradient of the function
def grad_f(x, y):
    return 2*x, 2*y

# Initialize
x, y = 4.0, 3.0  # Starting point
lr = 0.1         # Learning rate
steps = 30

# Store the path
path_x, path_y, path_z = [x], [y], [f(x, y)]

# Gradient descent loop
for _ in range(steps):
    dx, dy = grad_f(x, y)
    x -= lr * dx
    y -= lr * dy
    path_x.append(x)
    path_y.append(y)
    path_z.append(f(x, y))

# Plot the surface and path
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)
ax.plot(path_x, path_y, path_z, color='red', marker='o')
ax.set_title("2D Gradient Descent Path")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Loss")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# === Function and Gradient ===
def f(x, y):
    return x**4 - x**2 + y**2

def grad_f(x, y):
    df_dx = 4 * x**3 - 2 * x
    df_dy = 2 * y
    return df_dx, df_dy

# === Gradient Descent ===
def gradient_descent(lr, epochs, x0=0.0, y0=1.4):
    x, y = x0, y0
    path = [(x, y, f(x, y))]
    
    for _ in range(epochs):
        dx, dy = grad_f(x, y)
        x -= lr * dx
        y -= lr * dy
        path.append((x, y, f(x, y)))
    
    return path

# === Contour Plot ===
def plot_contour(path, title):
    x_vals, y_vals, _ = zip(*path)
    X = np.linspace(-2, 2, 400)
    Y = np.linspace(-2, 2, 400)
    X, Y = np.meshgrid(X, Y)
    Z = f(X, Y)

    plt.figure(figsize=(7,6))
    plt.contour(X, Y, Z, levels=50, cmap='viridis')
    plt.plot(x_vals, y_vals, marker='o', color='red', linewidth=2, label='Path')
    plt.title(f"Contour: {title}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# === 3D Surface Plot ===
def plot_3d(path, title):
    x_vals, y_vals, z_vals = zip(*path)
    X = np.linspace(-2, 2, 100)
    Y = np.linspace(-2, 2, 100)
    X, Y = np.meshgrid(X, Y)
    Z = f(X, Y)

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, alpha=0.6, cmap='viridis')
    ax.plot(x_vals, y_vals, z_vals, color='red', marker='o', label='Path')
    ax.set_title(f"3D Surface: {title}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("f(x, y)")
    ax.legend()
    plt.show()

# === Run Experiments ===

# 1. Initial guess [0, 1.4], lr=0.01, 1000 epochs
path1 = gradient_descent(lr=0.01, epochs=1000)
plot_contour(path1, "LR=0.01, Epochs=1000")
plot_3d(path1, "LR=0.01, Epochs=1000")

# 2. Same start, increase epochs to 10,000
path2 = gradient_descent(lr=0.01, epochs=10000)
plot_contour(path2, "LR=0.01, Epochs=10000")
plot_3d(path2, "LR=0.01, Epochs=10000")

# 3. Try LR = 1
path3 = gradient_descent(lr=1, epochs=1000)
plot_contour(path3, "LR=1.0, Epochs=1000")
plot_3d(path3, "LR=1.0, Epochs=1000")

# 4. Try LR = 0.5
path4 = gradient_descent(lr=0.5, epochs=1000)
plot_contour(path4, "LR=0.5, Epochs=1000")
plot_3d(path4, "LR=0.5, Epochs=1000")

# 5. Try LR = 0.00001
path5 = gradient_descent(lr=0.00001, epochs=1000)
plot_contour(path5, "LR=0.00001, Epochs=1000")
plot_3d(path5, "LR=0.00001, Epochs=1000")
```
